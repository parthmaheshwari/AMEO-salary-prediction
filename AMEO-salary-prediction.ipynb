```python
ameo_df['CumulativeDomain'] = ameo_df['ElectronicsAndSemicon'] + ameo_df['ComputerScience'] + ameo_df['MechanicalEngg'] + ameo_df['ElectricalEngg'] + ameo_df['TelecomEngg'] + ameo_df['CivilEngg']
```

```python
ameo_df['CumulativeDomain'].value_counts()
```

```python
import numpy as np #made a new column 'AGE' (as of 2015), DOB can be dropped now
import datetime as DT
import pandas as pd
now = pd.Timestamp(DT.datetime.now())
ameo_df['DOB'] = ameo_df['DOB'].where(ameo_df['DOB'] < now, ameo_df['DOB'] -  np.timedelta64(100, 'Y'))   # 2
ameo_df['AGE'] = (now - ameo_df['DOB']).astype('<m8[Y]') - 3
```

```python
#graduation Year has slight negative correlation with salary 
#the numeric value of year of graduation increases after each passing year 
#Creating new Column "YearAfterGrad" to count years passed after graduation of the applicant.
ameo_df['YearAfterGrad'] = 2015 - ameo_df['GraduationYear']
```

```python
#Improvement in class 12th compared to class 10th
ameo_df['Improv12'] = ((ameo_df['12percentage'] - ameo_df['10percentage'])/ameo_df['10percentage'])*100
```

```python
ameo_df['ImprovGrad'] = ((ameo_df['collegeGPA'] - ameo_df['12percentage'])/ameo_df['12percentage'])*100
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
import numpy as np
import pandas as pd

temp = np.array(['Gender','Degree','12board','Specialization'])
num_ohc_cols = pd.Series(temp)
# Copy of the data
data_ohc = ameo_df.copy()

# The encoders
le = LabelEncoder()
ohc = OneHotEncoder()

for col in num_ohc_cols:
    
    # Integer encode the string categories
    dat = le.fit_transform(data_ohc[col]).astype(np.int)
    
    # Remove the original column from the dataframe
    data_ohc = data_ohc.drop(col, axis=1)

    # One hot encode the data--this returns a sparse array
    new_dat = ohc.fit_transform(dat.reshape(-1,1))

    # Create unique column names
    n_cols = new_dat.shape[1]
    col_names = ['_'.join([col, str(x)]) for x in range(n_cols)]

    # Create the new dataframe
    new_df = pd.DataFrame(new_dat.toarray(), 
                          index=data_ohc.index, 
                          columns=col_names)

    # Append the new data to the dataframe
    data_ohc = pd.concat([data_ohc, new_df], axis=1)
```

```python
x = data_ohc.iloc[:, 1:].values
y = data_ohc["Salary"].values
x
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=11111)
```

```python
from sklearn.preprocessing import StandardScaler
normalizer = StandardScaler()
X_train_norm = normalizer.fit_transform(X_train)
X_test_norm = normalizer.transform(X_test)
```

```python
from keras.models  import Sequential, K
from keras.layers import Input, Dense, Flatten, Dropout, BatchNormalization
from keras.optimizers import Adam, SGD, RMSprop
```

```python
model_1 = Sequential([
    Dense(29, input_shape=(29,), activation="relu"),
    Dense(18, activation="relu"),
    Dense(12, activation="relu"),
    Dense(6, activation="relu"),
    Dense(1)
])
model_1.summary()
```

```python
model_1.compile(loss='mean_absolute_error', optimizer='adam')
run_hist_1 = model_1.fit(X_train_norm, y_train, validation_data=(X_test_norm, y_test), epochs=270)
```

```python
y_pred = model_1.predict(X_test_norm)
```

```python
for i in range(len(X_test_norm)):
	print("Actual=%s, Predicted=%s" % (y_test[i], y_pred[i]))
```
